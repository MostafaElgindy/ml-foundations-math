# Introduction to Linear Algebra

**Algebra** is a branch of mathematics dealing with symbols and the rules for manipulating those symbols.

$\text{Ex: } x + 2 = 5\rightarrow x=5-2=3$

**Linear Algebra** is a subfield of algebra that focuses on vector spaces and linear mappings between these spaces. It involves the study of vectors, matrices, and systems of linear equations.

General form of a linear equation:

$$
a_1x_1 + a_2x_2 + \ldots + a_nx_n = b\Rightarrow \begin{bmatrix}
a_{11} & a_{12} & \ldots & a_{1n} \\ a_{21} & a_{22} & \ldots & a_{2n} \\ \vdots & \vdots & \ddots & \vdots \\ a_{m1} & a_{m2} & \ldots & a_{mn}
\end{bmatrix}
\begin{bmatrix}
x_1 \\ x_2 \\ \vdots \\ x_n
\end{bmatrix}
=
\begin{bmatrix}
b_1 \\ b_2 \\ \vdots \\ b_m
\end{bmatrix}
$$

so, we can say $b$ is the price of the house, $x$ is the vector of features, and $a$ is the matrix of coefficients.





```python
import numpy as np
import matplotlib.pyplot as plt
```

EX: We have two cars, a red car and a blue car. The red car is moving at a speed of 2.5 km/s and the blue car is moving at a speed of 3 km/s. The blue car starts moving 5 seconds after the red car.

- red car: $d_r = 2.5t$
- blue car: $d_b = 3(t - 5)$

now let's equate the distance travelled by both cars at a given time $t$ and plot the results, so the equation becomes:
$$
2.5t = 3(t - 5)\Rightarrow 2.5t = 3t - 15 \Rightarrow 0.5t = 15 \Rightarrow t = 30
$$



```python
t = np.linspace(0, 40, 1000)  # time from 0 to 40 seconds, with 1000 points

d_r = 2.5 * t # distance travelled by the red car
d_b = 3 * (t - 5) # distance travelled by the blue car

plt.title("Car Distance Over Time")
plt.xlabel("Time (m)")
plt.ylabel("Distance (km)")

plt.xlim([0, 40])
plt.ylim([0, 100])

plt.plot(t, d_r, label="Red Car", color='red')
plt.plot(t, d_b, label="Blue Car", color='blue')

plt.axvline(30, color='green', linestyle='--') # vertical line at t=30 seconds
plt.axhline((2.5*30), color='green', linestyle='--') # horizontal line at the distance travelled by both cars at t=30 seconds

plt.legend()
```

The last example they have different speeds or slopes, but they meet at a point, which is the solution to the equation.

but what if they have the same speed? let's see the example below:

```python
t = np.linspace(0, 40, 1000)  # time from 0 to 40 seconds, with 1000 points

d_r = 2.5 * t # distance travelled by the red car
d_b = 2.5 * (t - 5) # distance travelled by the blue car, same speed but starts 5 seconds later

plt.title("Car Distance Over Time")
plt.xlabel("Time (m)")
plt.ylabel("Distance (km)")

plt.xlim([0, 40])
plt.ylim([0, 100])

plt.plot(t, d_r, label="Red Car", color='red')
plt.plot(t, d_b, label="Blue Car", color='blue')

plt.legend()
```

or if they have the same speed with the same starting points, like the example below:

```python
t = np.linspace(0, 40, 1000)  # time from 0 to 40 seconds, with 1000 points

d_r = 2.5 * t # distance travelled by the red car
d_b = 2.5 * t # distance travelled by the blue car, same speed and same starting point

plt.title("Car Distance Over Time")
plt.xlabel("Time (m)")
plt.ylabel("Distance (km)")

plt.xlim([0, 40])
plt.ylim([0, 100])

plt.plot(t, d_r, label="Red Car", color='red')
plt.plot(t, d_b, label="Blue Car", color='blue')

plt.legend()
```

```python
t = np.linspace(0, 40, 1000)  # Time from 0 to 40 seconds

# Scenario 1: Blue car is faster (3 km/s) but starts 5s later
d_r1 = 2.5 * t
d_b1 = 3 * (t - 5)

# Scenario 2: Blue car same speed (2.5 km/s) but starts 5s later
d_r2 = 2.5 * t
d_b2 = 2.5 * (t - 5)

# Scenario 3: Both cars same speed (2.5 km/s) and same start time
d_r3 = 2.5 * t
d_b3 = 2.5 * t

# Create a figure with 3 subplots (1 row, 3 columns)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Comparison of Car Distance Over Time (Three Scenarios)", fontsize=14)

# Common settings for all subplots
for ax in (ax1, ax2, ax3):
    ax.set_xlim([0, 40])
    ax.set_ylim([0, 100])
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Distance (km)")
    ax.grid(True, linestyle='--', alpha=0.5)

# --- Scenario 1 Plot ---
ax1.set_title("Scenario 1: Blue Car Faster (3 km/s), 5s Delay")
ax1.plot(t, d_r1, color='red', label='Red Car (2.5 km/s)')
ax1.plot(t, d_b1, color='blue', label='Blue Car (3 km/s, starts at t=5)')
ax1.axvline(30, color='green', linestyle='--', alpha=0.7, label='Intersection (t=30s)')
ax1.axhline(2.5*30, color='green', linestyle='--', alpha=0.7)
ax1.legend()

# --- Scenario 2 Plot ---
ax2.set_title("Scenario 2: Same Speed (2.5 km/s), Blue 5s Delay")
ax2.plot(t, d_r2, color='red', label='Red Car (2.5 km/s)')
ax2.plot(t, d_b2, color='blue', label='Blue Car (starts at t=5)')
ax2.legend()

# --- Scenario 3 Plot ---
ax3.set_title("Scenario 3: Same Speed (2.5 km/s), Same Start")
ax3.plot(t, d_r3, color='red', label='Red Car (2.5 km/s)')
ax3.plot(t, d_b3, color='blue', label='Blue Car (no delay)')
ax3.legend()

plt.tight_layout()
plt.show()
```

- In the second case, they never meet, which means there is no solution to the equation.

### Tensors
**Tensors** are a generalization of scalars, vectors, and matrices. They can be thought of as multi-dimensional arrays that can represent data in various dimensions.

- A **scalar** is a single number (0D tensor). $\text{Ex: } 5$
- A **vector** is a one-dimensional array of numbers (1D tensor). $\text{Ex: } \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}$
- A **matrix** is a two-dimensional array of numbers (2D tensor). $\text{Ex: } \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$
- A **tensor** can have three or more dimensions (3D tensor and beyond). $\text{Ex: } \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}$

- **Tensor**: A tensor is a multi-dimensional array that can hold data of any type, such as integers, floats, or strings. It is the fundamental data structure in TensorFlow and is used to represent the input and output of the model.

##### Vectors

A vector is a one-dimensional array of numbers. It can be represented as a row or a column.

**Vector transposition** is the operation of flipping a vector from a row to a column or vice versa. This is often used in linear algebra to change the orientation of a vector.

$$ \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}^T \Rightarrow \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} $$

**Vector Norms**:

- **L1 Norm** (Manhattan distance): The sum of the absolute values of the vector's components. $\|x\|_1 = |x_1| + |x_2| + \ldots + |x_n|$
- **L2 Norm** (Euclidean distance): The square root of the sum of the squares of the vector's components. $\|x\|_2 = \sqrt{x_1^2 + x_2^2 + \ldots + x_n^2}$
- **Infinity Norm**: The maximum absolute value of the vector's components. $\|x\|_\infty = \max(|x_1|, |x_2|, \ldots, |x_n|)$

and this is the vector norm in general:
$$
\|x\|_p = \left( |x_1|^p + |x_2|^p + \ldots + |x_n|^p \right)^{1/p}
$$

where $p$ is a positive integer.



```python
x = np.array([1, 2, 3])
print(f"Vector: {x}, Shape: {x.shape}")

x_transpose = x.reshape(-1, 1)  # Reshape to a column vector
print(f"Vector:\n{x_transpose}\nShape: {x_transpose.shape}")
```

**Basis Vectors**

In linear algebra, basis vectors are a set of vectors that span a vector space. They are linearly independent and can be used to represent any vector in that space.

$$ \text{Ex: } \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \begin{bmatrix} 0 \\ 1 \end{bmatrix} $$

**Orthogonal Vectors**

Two vectors are orthogonal if their dot product is zero. This means they are perpendicular to each other in the vector space.

$$ \text{Ex: } \begin{bmatrix} 1 \\ 0 \end{bmatrix} \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix} = 0 $$

**Orthonormal Vectors**

Two vectors are orthonormal if they are orthogonal and each vector has a norm of 1. This means they are both perpendicular and unit vectors.

$$ \text{Ex: } \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \begin{bmatrix} 0 \\ 1 \end{bmatrix} $$
